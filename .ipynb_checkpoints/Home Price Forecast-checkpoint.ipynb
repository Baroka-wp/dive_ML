```python
import pandas as pd
import numpy as np

# import and extract the explanatory variables GrLivArea and YearBuilt, and the objective variable SalePrice, respectively.
Data = pd.read_csv('train.csv')[["GrLivArea","YearBuilt", "SalePrice"]]

Data
```

## [Problem 2] Estimation and evaluation by linear regression

#### SK Learn — Diviser nos données

```python
# import Sk Learn train test split
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# varible descriptive
X = data[["GrLivArea","YearBuilt"]]

# variable explicative
y = data["SalePrice"]

#  Assign variables to capture train test split output
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Pretraitement et standardisation

scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

#### SK Learn — Entraîner notre modèle

```python
from sklearn.linear_model import LinearRegression

#instanciation et ajustement du modele aux donees d'entrainement
linreg = LinearRegression()

#Fit model to training data
linreg.fit(X_train, y_train)
```

#### Faire des prédictions basées sur votre modèle

```python
y_pred = linreg.predict(X_test)
```

#### évaluation du modèle

https://towardsdatascience.com/linear-regression-in-python-9a1f5f000606

Trois métriques principales sont utilisées pour évaluer les modèles linéaires. Ce sont : l'erreur absolue moyenne (MAE), l'erreur quadratique moyenne (MSE) ou l'erreur quadratique moyenne (RMSE).

MAE : Le plus simple à comprendre. Représente l'erreur moyenne

MSE : Similaire à MAE mais le bruit est exagéré et les erreurs plus importantes sont « punies ». Il est plus difficile à interpréter que le MAE car il n'est pas en unités de base, cependant, il est généralement plus populaire.

RMSE : métrique la plus populaire, similaire à MSE, cependant, le résultat est une racine carrée pour le rendre plus interprétable car il est en unités de base. Il est recommandé d'utiliser le RMSE comme métrique principale pour interpréter votre modèle.

Ci-dessous, vous pouvez voir comment calculer chaque métrique. Tous nécessitent deux listes comme paramètres, l'une étant vos valeurs prédites et l'autre étant les vraies valeurs

```python
#For regression problems, it is common to use Mean Squared Error (MSE) as an index value for evaluation.
from sklearn.metrics import mean_squared_error

print("MSE: {}".format(mean_squared_error(y_test, y_pred)))
```

## [Problem 3] Comparison of methods


```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

#### LinearRegression

```python
linreg = LinearRegression().fit(X_train, y_train)
y_pred_LR = linreg.predict(X_test)
visualization(y_pred_LR)
```

#### SVM 

```python
model_SVM = make_pipeline(StandardScaler(), SVC(gamma='auto')).fit(X_train, y_train)
y_pred_SVM = model_SVM.predict(X_test)
visualization(y_pred_SVM)
```

#### Decision tree

```python
model_DT = DecisionTreeClassifier(random_state=0).fit(X_train, y_train)
y_pred_DT = model_DT.predict(X_test)
visualization(y_pred_DT)
```

#### Random forest

```python
model_RF = RandomForestClassifier(max_depth=2, random_state=0).fit(X_train, y_train)
y_pred_RF = model_RF.predict(X_test)
visualization(y_pred_RF)
```
