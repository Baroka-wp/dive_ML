### [Problem 1] Creating random numbers


```python
import numpy as np

np.random.seed(0) # Fixing random state for reproducibility

mean = [-3, 0]
cov = [[1., .8], [.8, 1.]]

a_list = np.random.multivariate_normal(mean, cov, (500))

print(a_list.shape)
```

```python
import matplotlib.pyplot as plt


x = a_list[:,0] # extract first column
y = a_list[:,1] # extract second column

plt.title("scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(y,x,c="b",alpha=0.5,label="factice grains")

plt.show()
```

### [Problem 3] Visualization by histogram

```python
plt.title("histogram x1")
plt.xlabel("x1")
plt.ylabel("frequency")
plt.hist(x,50, density=True, facecolor='g', alpha=1) 
plt.xlim(-5, 2)

plt.show()
```

```python
plt.title("histogram x2")
plt.xlabel("x2")
plt.ylabel("frequency")
plt.hist(y,50, density=True, facecolor='b', alpha=1)
plt.xlim(-5, 2)
plt.show()
```

### [Probl√®me 4] add data

```python
import numpy as np

np.random.seed(2)# Fixing random state for reproducibility

mean = [0, -3]
cov = [[1.0, 0.8], [0.8, 1.0]]

b_list = np.random.multivariate_normal(mean, cov, (500))

print(b_list.shape)
```

```python
#scatter diagram
import matplotlib.pyplot as plt

a = b_list[:,0] # extract first column
b = b_list[:,1] # extract second column

plt.title("scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(x,y,c="b",alpha=0.5,label="0")
plt.scatter(a,b,c="r",alpha=0.5,label="1")
plt.legend(loc='upper left')

plt.show()
```

```python
# histogram x1, a

plt.title("histogram x1")
plt.xlabel("x1")
plt.ylabel("frequency")
plt.hist(x,50, density=True, facecolor='g', alpha=1)
plt.hist(a,50, density=True, facecolor='r', alpha=1)
plt.xlim(-5, 2)

plt.show()
```

```python
# histogram x2, b

plt.title("histogram x2")
plt.xlabel("x2")
plt.ylabel("frequency")
plt.hist(y,50, density=True, facecolor='g', alpha=1)
plt.hist(b,50, density=True, facecolor='r', alpha=1)
plt.xlim(-5, 2)

plt.show()
```

### [Problem 5] Data combination

```python
# using np.concatenate()
c_list = np.concatenate((a_list, b_list))
print(a_list.shape)
print(c_list.shape)
```

```python
#using np.vstack()

d_list = np.vstack((a_list, b_list))
print(d_list.shape)
```

```python
# create labeling array (l_ndarray)

l_ndarray = np.concatenate((np.full(500, 0),np.full(500, 1))).reshape(1000,1)

print(l_ndarray.shape)
```

```python
f_ndarray = np.concatenate((d_list, l_ndarray), axis=1)
print(f_ndarray)
```
