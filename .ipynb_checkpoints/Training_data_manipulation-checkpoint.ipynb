### Acces indicés aux donnees d'un DataFrame

On peut accéder aux valeurs du DataFrame via des indices ou plages d'indice. La structure se comporte alors comme une matrice. La cellule en haut et à gauche est de coordonnées (0,0).


Il y a différentes manières de le faire, l'utilisation de .iloc[,] constitue une des solutions les plus simples. N'oublions pas que Shape permet d'obtenir les dimensions (lignes et colonnes) du DataFrame.

```python
# acces a la valeur (0,0)

df.iloc[0,0] #(cellule : ligne 0 et colonne 0)
```

```python
# derniere ligne, premiere colonne

df.iloc[-1,0]
```

```python
#valeur située en dernière ligne, première colonne 
#shape[0] renvoie le nombre de lignes (première dimension) 
#il faut réduire de -1 parce le premier indice est égal à 0 
#sinon on déborde

df.iloc[df.shape[0]-1,0]
```

```python
# 5 premiere valeur de toutes les colonnes
# ligne => (0:5) 0 a 5 non inclus
# colonnes = : (toutes les colonnes)

df.iloc[0:5,:]
#df.iloc[:5,:]
```

```python
# selectionner les 5 dernieres lignes et toutes les colonnes

df.iloc[-5:,:]
```

```python
# 5 premieres lignes et 2 permieres colonnes

df.iloc[0:5,0:2]
```

```python
# 5 premieres lignes et les colonnes 0, 2, 4
# ( c'est une liste d'indices)

df.iloc[0:5,[0,2,4]]

df.iloc[0:5, 0:5:2] 
# 5 premieres lignes et les colonnes de 0 a 5(non inclus) en sautant deux par deux
```

### Restrictions avec les conditions - Les requêtes

Nous pouvons isoler les sous-ensembles d'observations répondant à des critères définis sur les champs. Nous utiliserons préférentiellement la méthode .loc[,] dans ce cadre.

```python
# liste des individu presentant des douleurs de type "A"

df.loc[df['type_douleur']=="A",:]

# => df['type_douleur']=="A" renvoie un vecteur boolean
# .loc[,] ne reprend que les valeurs True
```

```python
# contabiliser les True et False

(df['type_douleur']=="A").value_counts()
 
```

```python
# liste des individus presentant une douleur de type "A" ou "B"

df.loc[df["type_douleur"].isin(["A","B"]),:]
```

Des opérateurs logiques permettent de combiner les conditions. Nous utilisons respectivement : & pour ET, | pour OU, et ~ pour la négation

```python
# liste des individus présentant une douleur de type A et angine == oui 

df.loc[(df['type_douleur']=="A") & (df['angine'] == "oui"),:]
 
```

```python
#on peut n'afficher qu'une partie des colonnes
#on définit la projection dans une liste
colonnes = ['age','sexe','coeur','taux_max']
#que l'on utilise en paramètre dans .loc[]
#pour la même restruction que précédemment
print(df.loc[(df['age'] < 45) & (df['sexe'] == "masculin") & (df['coeur'] == "presence"),colonnes])
```

### Calculs récapitulatifs - Croisement des variables

A la manière des tableaux croisés dynamiques (TCD) d'Excel, nous pouvons procéder à des croisements et opérer des calculs récapitulatifs, qui vont du comptage simple aux calculs statistiques mettent en jeu d'autres variables.

```python
#fréquences selon sexe et coeur - cela ressemnble à table() de R
#voir : http://pandas.pydata.org/pandas-docs/stable/generated/pandas.crosstab.ht ml


pd.crosstab(df['sexe'],df['coeur'])
 
```

```python
# nous pouvons demander un post-traitement, à la différence de table() de R 
# par exemple, un pourcentage en ligne 

pd.crosstab(df['sexe'],df['coeur'],normalize='index')
```

```python
#nous pouvons aussi insérer un champ calculé, comme avec les TCD d'Excel
#ex. moyenne d'âge selon le sexe et la maladie
#nous utilisons la fonction mean() de la classe Series de la librairie Pandas #cela nous rapproche plus des fonctions tapply() ou aggregate() de R 

print(pd.crosstab(df['sexe'],df['coeur'],values=df['age'],aggfunc=pd.Series.mean))
```

```python
print(df.pivot_table(index=['sexe'],columns=['coeur'],values=['age'],aggfunc=pd.Series.mean))
```

```python
# multiplier les criteres 
pd.crosstab([df["sexe"],df["sucre"]],df["coeur"])
```

```python
# afficher les proportions

pd.crosstab([df["sexe"],df["sucre"]],df["coeur"], normalize="index")
```


L'utilisation de groupby() permet d'accéder aux sous-DataFrame associés à chaque item de la variable de regroupement. Il est dès lors possible d'appliquer explicitement d'autres traitements sur ces sous-ensembles de données.

```python
# scission des donnees selon le sexe

g = df.groupby("sexe")

#Recuperer les donnees des hommes uniquement
g.get_group("masculin")
```

```python
# calculer la moyenne d'age des hommes 

g.get_group('masculin')['age'].mean()
```

```python
g[['age','depression']].agg([pd.Series.mean,pd.Series.std])
 
```

```python
# on peut iterer sur les groupe

for group in g:
    print(group[0])
    print(group[1]["age"].mean())    
    #print(pd.Series.mean(group[1]["age"]))
```
