## [Problème 4] Créer un fichier pour apprendre


```python
cd /content/drive/MyDrive/ObjectDetection
```

```python
ls
```

```python
import pandas as pd
annotation_df = pd.read_csv('annotation.txt',header=None)
n_sample, n_col = annotation_df.shape
print(n_sample,n_col)
annotation_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
annotation_df.iloc[:,5] = le.fit_transform(annotation_df.iloc[:,5])
annotation_df.tail()
```

```python
path = 'annotation.txt'
out_path = 'train.txt'

with open(path) as f:
  lines = f.readline() 
  for i in range(n_sample):
    split_line = lines.split(',') 
    image_path = split_line[0]
    split_line[0] = './' + image_path 
    split_line[-1] = str(annotation_df.iloc[i,5]) + '\n'  

    with open(out_path, mode='a') as out_f:
      join_line = ','.join(split_line)  
      join_line = join_line.replace('.jpg,','.jpg ')  
      out_f.write(join_line)
```

```python
pd.Series(le).to_csv('class_label.txt',index=False,header=False)
```
